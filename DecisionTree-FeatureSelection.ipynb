# Decision Tree - Feature Selection

```python
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
```

```python
boston = load_boston()
```

```python
X = boston.data
```

```python
y = boston.target
```

```python
tree_clf = DecisionTreeRegressor()
```

```python
tree_clf.fit(X,y)
```

```python
export_graphviz(tree_clf, out_file = 'boston.dot', 
                feature_names = boston.feature_names,
                rounded = True,
                filled = True
               )
```

```python
for line in open('boston.dot'):
    print(line)
```

## Feature Importance

```python
sorted({n: w for n, w in zip(boston.feature_names, tree_clf.feature_importances_)}.items(), key = lambda x: x[1], reverse = True)
```
