# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def func(x):
    return 10*x**2 + 37*x +9
```

```python
def gradient(x):
    return 20*x + 37
```

```python
x = np.linspace(-10,10)
plt.plot(x, func(x))
```

```python
steps = []

x_optimal = random.choice(x)
x_optimal
```

$$ x' = x + (-1)*\frac{\partial{y}}{\partial{x}}\alpha$$

```python
alpha = 1e-2

for i in range(300):
    x_optimal = x_optimal + (-1) * gradient(x_optimal) * alpha
    steps.append(x_optimal)
```

```python
steps
```

```python
for s in steps:
    print(s, func(s))
```
